## get duplicate table

```python
t = tanimoto(df.copy(),target_col = "IC50")
```

```python
t = t.query('SimilarityScore == 1')
```

```python
dup = pd.DataFrame(t.groupby("ID1").ID2.unique()).reset_index()
```

```python
ID1 = dup.ID1.tolist()
```

```python
remove = []
for id2 in d.ID2:
    for id1 in ID1:
        if id1 in id2:
            remove.append(ID1.index(id1))
```

```python
remove
```

```python
dup = dup[~dup.index.isin(remove)].reset_index(drop=True)
```

```python
dup2 = pd.DataFrame(dup.apply(lambda x: [x["ID1"]] + [r for r in x["ID2"]], axis=1))
```

```python
for i in ['IC50','erk_IC50','Kd']:
    dup2[i] = dup2[0].apply(lambda lis: [df[df.ID == l][i].values.item() for l in lis])
    dup2[f'{i}_mean'] = dup2[i].apply(lambda lis: round(np.mean([x for x in lis if not np.isnan(x)]),2))
```

```python
dup2
```

```python
dup2['ID'] = dup2[0].apply(lambda lis: lis[-1])
```

```python
dup2['SMILES'] = dup2['ID'].apply(lambda i: df[df.ID == i].SMILES.values.item())
```

```python
dup3 = dup2[['ID','SMILES','IC50_mean','erk_IC50_mean','Kd_mean']].rename(columns = {'IC50_mean':'IC50','erk_IC50_mean':'erk_IC50','Kd_mean':'Kd'})
```
