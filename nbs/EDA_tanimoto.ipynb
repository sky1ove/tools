```python
from fastbook import *
import seaborn as sns

from tools.toolbox import *
from tools.dataset import Data
```

```python
sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
sns.set_context('notebook')
sns.set_style("ticks")
```

```python
df = Data.get_g12d()

cols = ['ID','SMILES','IC50']

df = df[cols]

df = df.dropna(subset= 'IC50').reset_index(drop=True)
```

```python
t = tanimoto(df, target_col = 'IC50')
```

```python
t['abs_log10'] = t.apply(lambda x: abs(np.log10(x['Target1']/x['Target2'])), axis=1)
```

```python
t.plot.scatter(x='SimilarityScore', y='abs_log10');
```

```python
# Create a scatter plot with Seaborn
sns.scatterplot(data=t,
                x='SimilarityScore',
                y='abs_log10',
                s=10)

plt.xlabel('Tanimoto Similarity Score')
plt.ylabel('|log10(value1/value2)|')

# Show the plot
plt.show()
```

```python
# Create a scatter plot with Seaborn
sns.kdeplot(data=t,
            x='SimilarityScore',
            y='abs_log10', 
            cmap='Blues', 
            shade=True)

plt.xlabel('Tanimoto Similarity Score')
plt.ylabel('|log10(value1/value2)|')

# Show the plot
plt.show()
```
